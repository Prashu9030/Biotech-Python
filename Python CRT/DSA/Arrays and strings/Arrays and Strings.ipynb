```python
#MODULAR DIVISION OF ARRAY
x=np.array([1,2,3,4,5])
y=np.array([5,6,7,8,9,])
print("Given x:",x)
print("Given y:",y)
c=x%y
print("After Division :",c)
```

Program to count the uppercase alphabets,lowercase alphabets,digits and symbols present in the input string.

```python
def count_characters(s):
    upper = lower = digit = symbol = 0

    for char in s:
        if char.isupper():
            upper += 1
        elif char.islower():
            lower += 1
        elif char.isdigit():
            digit += 1
        else:
            symbol += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)
    print("Digits:", digit)
    print("Symbols:", symbol)

# Input from user
input_string = input("Enter a string: ")
count_characters(input_string)
```
